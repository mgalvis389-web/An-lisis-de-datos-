```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar archivo CSV
df = pd.read_csv("nacimientossclase.csv",encoding="latin1")

# Información básica
print("Información del dataset")
print(df.info())
print("Primeras filas del dataset")
print(df.head())
print("Resumen estadístico del dataset")
print(df.describe())

```

```python
from scipy.stats import shapiro, kstest, anderson, jarque_bera
import statsmodels.api as sm
import matplotlib.pyplot as plt

numerical_cols = ['EDAD MADRE', 'EDAD PADRE', 'PESO (Gramos)']

print("Normality Tests:")
for col in numerical_cols:
    print(f"\n--- {col} ---")

    # Shapiro-Wilk Test
    shapiro_test = shapiro(df_cleaned_iqr[col])
    print(f"Shapiro-Wilk Test: Statistic={shapiro_test.statistic:.4f}, p-value={shapiro_test.pvalue:.4f}")

    # Kolmogorov-Smirnov Test (vs. Normal)
    # Note: Requires specifying the distribution to test against. Testing against normal distribution.
    # This test is sensitive to the distribution's parameters (mean and std dev),
    # which are estimated from the data.
    ks_test = kstest(df_cleaned_iqr[col], cdf='norm')
    print(f"Kolmogorov-Smirnov Test (vs. Normal): Statistic={ks_test.statistic:.4f}, p-value={ks_test.pvalue:.4f}")

    # Anderson-Darling Test
    anderson_test = anderson(df_cleaned_iqr[col])
    print(f"Anderson-Darling Test: Statistic={anderson_test.statistic:.4f}")
    for i in range(len(anderson_test.critical_values)):
        sl, cv = anderson_test.significance_level[i], anderson_test.critical_values[i]
        if anderson_test.statistic < cv:
            print(f"  At {sl}% significance level, data looks normal (critical value = {cv:.4f})")
        else:
            print(f"  At {sl}% significance level, data does not look normal (critical value = {cv:.4f})")

    # Jarque-Bera Test
    # Note: Jarque-Bera is more suitable for larger datasets and tests for skewness and kurtosis
    jb_test = jarque_bera(df_cleaned_iqr[col])
    print(f"Jarque-Bera Test: Statistic={jb_test.statistic:.4f}, p-value={jb_test.pvalue:.4f}")

    # QQ Plot
    plt.figure(figsize=(8, 6))
    sm.qqplot(df_cleaned_iqr[col], line='s') # line='s' plots a 45-degree line
    plt.title(f'QQ Plot of {col}')
    plt.show()
```
