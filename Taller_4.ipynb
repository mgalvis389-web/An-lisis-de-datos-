1. Se desea estimar la altura promedio de estudiantes universitarios. En una muestra de 50 estudiantes,
se obtuvo una media de 170 cm. Se sabe que la desviación estándar poblacional es de 10 cm.
Construye un intervalo de confianza del 95% para la media poblacional.

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# -----------------------------------------------------------------------------
# Problema: Intervalo de confianza para la media poblacional
# -----------------------------------------------------------------------------

# --- 1. Parámetros dados en el problema ---
n = 50  # Tamaño de la muestra
x_bar = 170  # Media de la muestra (cm)
sigma = 10  # Desviación estándar poblacional (cm)
nivel_confianza = 0.95  # Nivel de confianza deseado

# -----------------------------------------------------------------------------
# --- 2. Cálculo del Intervalo de Confianza ---
# -----------------------------------------------------------------------------

# La fórmula para el intervalo de confianza de la media poblacional (con sigma conocida) es:
# IC = x_bar +/- Z * (sigma / sqrt(n))

# --- Paso 2.1: Encontrar el valor crítico Z ---
# El nivel de significancia (alfa) es 1 - nivel_confianza
alfa = 1 - nivel_confianza
# El valor Z_alfa/2 se obtiene de la distribución normal estándar.
# Usamos scipy.stats.norm.ppf() para encontrar el percentil (punto Z)
# correspondiente a la probabilidad (1 - alfa/2).
z_critico = norm.ppf(1 - alfa / 2)

print(f"Nivel de confianza: {nivel_confianza * 100}%")
print(f"Nivel de significancia (alfa): {alfa}")
print(f"Valor crítico Z: {z_critico:.4f}")

# --- Paso 2.2: Calcular el error estándar (sigma_x_bar) ---
# El error estándar es la desviación estándar poblacional dividida por la raíz cuadrada del tamaño de la muestra.
error_estandar = sigma / np.sqrt(n)

print(f"Error estándar: {error_estandar:.4f}")

# --- Paso 2.3: Calcular el margen de error ---
# El margen de error es el valor crítico Z multiplicado por el error estándar.
margen_error = z_critico * error_estandar

print(f"Margen de error: {margen_error:.4f} cm")

# --- Paso 2.4: Construir el intervalo de confianza ---
# El límite inferior es la media muestral menos el margen de error.
limite_inferior = x_bar - margen_error
# El límite superior es la media muestral más el margen de error.
limite_superior = x_bar + margen_error

print(f"\nIntervalo de confianza del {nivel_confianza * 100}%:")
print(f"[{limite_inferior:.2f} cm, {limite_superior:.2f} cm]")

# -----------------------------------------------------------------------------
# --- 3. Graficación de la Distribución ---
# -----------------------------------------------------------------------------

# --- Paso 3.1: Definir la distribución normal para la media muestral ---
# La distribución muestral de la media se aproxima a una distribución normal
# con media mu (media poblacional) y desviación estándar igual al error estándar.
# Para la gráfica, usaremos la media muestral x_bar como el centro.
mu_distribucion = x_bar
sigma_distribucion = error_estandar

# Generar puntos en el eje x para la curva normal
x = np.linspace(mu_distribucion - 4 * sigma_distribucion, mu_distribucion + 4 * sigma_distribucion, 200)
# Calcular los valores de la función de densidad de probabilidad (PDF)
y = norm.pdf(x, mu_distribucion, sigma_distribucion)

# --- Paso 3.2: Configurar la gráfica ---
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Distribución Muestral de la Media', color='blue')
plt.title('Intervalo de Confianza del 95% para la Altura Promedio')
plt.xlabel('Altura (cm)')
plt.ylabel('Densidad de Probabilidad')
plt.grid(True)

# --- Paso 3.3: Graficar el intervalo de confianza ---
# Rellenar el área bajo la curva que representa el 95% de confianza.
# El área sombreada se extiende desde el límite inferior hasta el superior.
x_fill = np.linspace(limite_inferior, limite_superior, 100)
y_fill = norm.pdf(x_fill, mu_distribucion, sigma_distribucion)
plt.fill_between(x_fill, y_fill, color='lightblue', alpha=0.6, label=f'Área del {nivel_confianza * 100}% de Confianza')

# --- Paso 3.4: Marcar la media muestral y los límites del intervalo ---
plt.axvline(x=x_bar, color='red', linestyle='--', label=f'Media Muestral ($\\bar{{x}}$) = {x_bar} cm')
plt.axvline(x=limite_inferior, color='green', linestyle=':', label=f'Límite Inferior = {limite_inferior:.2f} cm')
plt.axvline(x=limite_superior, color='green', linestyle=':', label=f'Límite Superior = {limite_superior:.2f} cm')

plt.legend()
plt.show()

# -----------------------------------------------------------------------------
# --- 4. Conclusión ---
# -----------------------------------------------------------------------------

print("\n--- Conclusión ---")
print(f"Con un nivel de confianza del 95%, se puede afirmar que la altura promedio de todos los")
print(f"estudiantes universitarios se encuentra en el rango de {limite_inferior:.2f} cm a {limite_superior:.2f} cm.")
print("Esto significa que si repitiéramos este proceso de muestreo y cálculo de intervalos")
print("muchas veces, aproximadamente el 95% de esos intervalos contendrían la verdadera media poblacional.")
```

2. Un examen nacional tiene una desviación estándar histórica de 15 puntos. En una muestra aleatoria
de 100 estudiantes, la calificación promedio fue de 75 puntos. Determina un intervalo de confianza
del 90% para la media poblacional.

```python
# Importación de librerías necesarias
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# -----------------------------------------------------------------------------
# Problema: Intervalo de confianza para la media poblacional de calificaciones
# -----------------------------------------------------------------------------

# --- 1. Definición de los parámetros del problema ---
n = 100  # Tamaño de la muestra de estudiantes
x_bar = 75  # Calificación promedio de la muestra
sigma = 15  # Desviación estándar poblacional histórica
nivel_confianza = 0.90  # Nivel de confianza deseado

# -----------------------------------------------------------------------------
# --- 2. Cálculo del Intervalo de Confianza ---
# -----------------------------------------------------------------------------

# La fórmula utilizada es la del intervalo de confianza para la media cuando la
# desviación estándar poblacional (sigma) es conocida:
# IC = x_bar +/- Z * (sigma / sqrt(n))

# --- Paso 2.1: Encontrar el valor crítico Z ---
# El nivel de significancia (alfa) se calcula como 1 - nivel_confianza.
alfa = 1 - nivel_confianza
# Usamos la función ppf (percent point function) de scipy.stats.norm
# para encontrar el valor Z que corresponde a la probabilidad (1 - alfa/2).
z_critico = norm.ppf(1 - alfa / 2)

print(f"Nivel de confianza: {nivel_confianza * 100}%")
print(f"Nivel de significancia (alfa): {alfa}")
print(f"Valor crítico Z para el 90% de confianza: {z_critico:.4f}")

# --- Paso 2.2: Calcular el error estándar ---
# El error estándar de la media muestral es sigma / sqrt(n).
error_estandar = sigma / np.sqrt(n)

print(f"Error estándar: {error_estandar:.4f}")

# --- Paso 2.3: Calcular el margen de error ---
# El margen de error es el valor crítico Z multiplicado por el error estándar.
margen_error = z_critico * error_estandar

print(f"Margen de error: {margen_error:.4f} puntos")

# --- Paso 2.4: Construir los límites del intervalo ---
# Límite inferior = media muestral - margen de error
limite_inferior = x_bar - margen_error
# Límite superior = media muestral + margen de error
limite_superior = x_bar + margen_error

print(f"\nIntervalo de confianza del {nivel_confianza * 100}%:")
print(f"[{limite_inferior:.2f}, {limite_superior:.2f}] puntos")

# -----------------------------------------------------------------------------
# --- 3. Graficación de la Distribución ---
# -----------------------------------------------------------------------------

# --- Paso 3.1: Definir la distribución normal para la gráfica ---
# La distribución muestral de la media se aproxima a una normal con media x_bar
# y desviación estándar igual al error estándar.
mu_distribucion = x_bar
sigma_distribucion = error_estandar

# Generar un rango de valores en el eje x para la gráfica
x = np.linspace(mu_distribucion - 4 * sigma_distribucion, mu_distribucion + 4 * sigma_distribucion, 200)
# Calcular la función de densidad de probabilidad (PDF) para cada punto x
y = norm.pdf(x, mu_distribucion, sigma_distribucion)

# --- Paso 3.2: Configurar y dibujar la gráfica ---
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Distribución Muestral de la Media', color='blue')
plt.title('Intervalo de Confianza del 90% para la Calificación Promedio')
plt.xlabel('Calificación')
plt.ylabel('Densidad de Probabilidad')
plt.grid(True)

# --- Paso 3.3: Rellenar el área del intervalo de confianza ---
# El área sombreada representa el 90% de confianza.
x_fill = np.linspace(limite_inferior, limite_superior, 100)
y_fill = norm.pdf(x_fill, mu_distribucion, sigma_distribucion)
plt.fill_between(x_fill, y_fill, color='lightblue', alpha=0.6, label=f'Área del {nivel_confianza * 100}% de Confianza')

# --- Paso 3.4: Marcar la media muestral y los límites del intervalo ---
plt.axvline(x=x_bar, color='red', linestyle='--', label=f'Media Muestral ($\\bar{{x}}$) = {x_bar}')
plt.axvline(x=limite_inferior, color='green', linestyle=':', label=f'Límite Inferior = {limite_inferior:.2f}')
plt.axvline(x=limite_superior, color='green', linestyle=':', label=f'Límite Superior = {limite_superior:.2f}')

plt.legend()
plt.show()

# -----------------------------------------------------------------------------
# --- 4. Conclusión ---
# -----------------------------------------------------------------------------

print("\n--- Conclusión ---")
print(f"Con un nivel de confianza del 90%, se puede concluir que la verdadera calificación")
print(f"promedio de todos los estudiantes que tomaron el examen nacional se encuentra entre")
print(f"{limite_inferior:.2f} y {limite_superior:.2f} puntos.")
print("Esto significa que si se seleccionaran repetidamente muestras de 100 estudiantes y se")
print("construyera un intervalo de confianza del 90% para cada muestra, aproximadamente")
print("el 90% de esos intervalos contendrían la verdadera media poblacional.")
```

```python
# Importación de librerías
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# -----------------------------------------------------------------------------
# Problema: Intervalo de confianza para el crecimiento de plantas con fertilizante
# -----------------------------------------------------------------------------

# --- 1. Definición de los parámetros del problema ---
# Tamaño de la muestra
n = 40
# Crecimiento promedio de la muestra (cm)
x_bar = 20
# Desviación estándar poblacional (cm)
sigma = 4
# Nivel de confianza deseado
nivel_confianza = 0.95

# -----------------------------------------------------------------------------
# --- 2. Cálculo del Intervalo de Confianza ---
# -----------------------------------------------------------------------------

# La fórmula utilizada es la del intervalo de confianza para la media cuando la
# desviación estándar poblacional (sigma) es conocida:
# IC = x_bar +/- Z * (sigma / sqrt(n))

# --- Paso 2.1: Encontrar el valor crítico Z ---
# El nivel de significancia (alfa) es 1 - nivel_confianza.
alfa = 1 - nivel_confianza
# Usamos scipy.stats.norm.ppf() para encontrar el valor Z correspondiente a la probabilidad (1 - alfa/2).
z_critico = norm.ppf(1 - alfa / 2)

print(f"Nivel de confianza: {nivel_confianza * 100}%")
print(f"Nivel de significancia (alfa): {alfa}")
print(f"Valor crítico Z para el 95% de confianza: {z_critico:.4f}")

# --- Paso 2.2: Calcular el error estándar ---
# El error estándar es la desviación estándar poblacional dividida por la raíz cuadrada del tamaño de la muestra.
error_estandar = sigma / np.sqrt(n)

print(f"Error estándar: {error_estandar:.4f}")

# --- Paso 2.3: Calcular el margen de error ---
# El margen de error es el valor crítico Z multiplicado por el error estándar.
margen_error = z_critico * error_estandar

print(f"Margen de error: {margen_error:.4f} cm")

# --- Paso 2.4: Construir los límites del intervalo ---
# Límite inferior = media muestral - margen de error
limite_inferior = x_bar - margen_error
# Límite superior = media muestral + margen de error
limite_superior = x_bar + margen_error

print(f"\nIntervalo de confianza del {nivel_confianza * 100}%:")
print(f"[{limite_inferior:.2f} cm, {limite_superior:.2f} cm]")

# -----------------------------------------------------------------------------
# --- 3. Graficación de la Distribución ---
# -----------------------------------------------------------------------------

# --- Paso 3.1: Definir la distribución normal para la gráfica ---
# La distribución muestral de la media se aproxima a una normal.
mu_distribucion = x_bar
sigma_distribucion = error_estandar

# Generar un rango de valores en el eje x para la curva normal
x = np.linspace(mu_distribucion - 4 * sigma_distribucion, mu_distribucion + 4 * sigma_distribucion, 200)
# Calcular los valores de la función de densidad de probabilidad (PDF)
y = norm.pdf(x, mu_distribucion, sigma_distribucion)

# --- Paso 3.2: Configurar la gráfica ---
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Distribución Muestral de la Media', color='blue')
plt.title('Intervalo de Confianza del 95% para el Crecimiento Promedio de Plantas')
plt.xlabel('Crecimiento (cm)')
plt.ylabel('Densidad de Probabilidad')
plt.grid(True)

# --- Paso 3.3: Rellenar el área del intervalo de confianza ---
# El área sombreada representa el 95% de confianza.
x_fill = np.linspace(limite_inferior, limite_superior, 100)
y_fill = norm.pdf(x_fill, mu_distribucion, sigma_distribucion)
plt.fill_between(x_fill, y_fill, color='lightblue', alpha=0.6, label=f'Área del {nivel_confianza * 100}% de Confianza')

# --- Paso 3.4: Marcar la media muestral y los límites del intervalo ---
plt.axvline(x=x_bar, color='red', linestyle='--', label=f'Media Muestral ($\\bar{{x}}$) = {x_bar} cm')
plt.axvline(x=limite_inferior, color='green', linestyle=':', label=f'Límite Inferior = {limite_inferior:.2f} cm')
plt.axvline(x=limite_superior, color='green', linestyle=':', label=f'Límite Superior = {limite_superior:.2f} cm')

plt.legend()
plt.show()

# -----------------------------------------------------------------------------
# --- 4. Conclusión ---
# -----------------------------------------------------------------------------

print("\n--- Conclusión ---")
print(f"Con un nivel de confianza del 95%, podemos afirmar que el verdadero crecimiento")
print(f"promedio de las plantas con el fertilizante se encuentra en el rango de")
print(f"{limite_inferior:.2f} cm a {limite_superior:.2f} cm.")
print("Esto significa que si se repitiera el experimento y se calcularan los intervalos")
print("de confianza de esta manera, el 95% de ellos contendrían la verdadera media poblacional.")
```

6. Una cadena de cafeterías desea estimar la temperatura promedio (°C) del café que sirve a sus clientes.
Se tomó una muestra aleatoria de 12 tazas donde la media fue de 84°C, y la desviación estándar
5°C. Con un nivel de confianza del 95% calcule el intervalo de confianza para la temperatura media
poblacional del café.

```python
# Importación de librerías necesarias
import numpy as np
from scipy.stats import t # Importar la distribución t de Student
import matplotlib.pyplot as plt

# -----------------------------------------------------------------------------
# Problema 6: Intervalo de confianza para la temperatura promedio del café
# -----------------------------------------------------------------------------

# --- 1. Definición de los parámetros del problema ---
n = 12  # Tamaño de la muestra de tazas de café
x_bar = 84  # Temperatura promedio de la muestra (°C)
s = 5  # Desviación estándar muestral (°C)
nivel_confianza = 0.95  # Nivel de confianza deseado

# -----------------------------------------------------------------------------
# --- 2. Cálculo del Intervalo de Confianza ---
# -----------------------------------------------------------------------------

# La fórmula para el intervalo de confianza de la media poblacional (con sigma desconocida y n pequeño) es:
# IC = x_bar +/- t * (s / sqrt(n))

# --- Paso 2.1: Encontrar el valor crítico t ---
# El nivel de significancia (alfa) es 1 - nivel_confianza.
alfa = 1 - nivel_confianza
# Los grados de libertad para la distribución t son n - 1.
grados_libertad = n - 1
# Usamos scipy.stats.t.ppf() para encontrar el valor t correspondiente a la probabilidad (1 - alfa/2)
# con los grados de libertad adecuados.
t_critico = t.ppf(1 - alfa / 2, grados_libertad)

print(f"Nivel de confianza: {nivel_confianza * 100}%")
print(f"Nivel de significancia (alfa): {alfa}")
print(f"Grados de libertad: {grados_libertad}")
print(f"Valor crítico t para el {nivel_confianza * 100}% de confianza: {t_critico:.4f}")

# --- Paso 2.2: Calcular el error estándar estimado (s_x_bar) ---
# El error estándar estimado es la desviación estándar muestral dividida por la raíz cuadrada del tamaño de la muestra.
error_estandar_estimado = s / np.sqrt(n)

print(f"Error estándar estimado: {error_estandar_estimado:.4f}")

# --- Paso 2.3: Calcular el margen de error ---
# El margen de error es el valor crítico t multiplicado por el error estándar estimado.
margen_error = t_critico * error_estandar_estimado

print(f"Margen de error: {margen_error:.4f} °C")

# --- Paso 2.4: Construir los límites del intervalo ---
# Límite inferior = media muestral - margen de error
limite_inferior = x_bar - margen_error
# Límite superior = media muestral + margen de error
limite_superior = x_bar + margen_error

print(f"\nIntervalo de confianza del {nivel_confianza * 100}%:")
print(f"[{limite_inferior:.2f} °C, {limite_superior:.2f} °C]")


# -----------------------------------------------------------------------------
# --- 3. Graficación de la Distribución ---
# -----------------------------------------------------------------------------

# Para la gráfica, usaremos la distribución t con grados de libertad n-1.
# Aunque la distribución t se centra en 0, la distribución muestral de la media
# se centra en la media muestral (x_bar).
mu_distribucion = x_bar
# La desviación estándar de la distribución muestral es el error estándar estimado.
sigma_distribucion = error_estandar_estimado

# Generar puntos en el eje x para la curva de la distribución t centrada en x_bar
# Usaremos un rango más amplio para visualizar mejor la forma de la distribución t
x = np.linspace(mu_distribucion - 4 * sigma_distribucion, mu_distribucion + 4 * sigma_distribucion, 200)
# Calcular los valores de la función de densidad de probabilidad (PDF) de la distribución t
# Es importante notar que norm.pdf aquí es para ilustrar la forma de campana centrada en x_bar
# La gráfica representa la distribución muestral de la media, que se modela con la t.
# Sin embargo, la función norm.pdf se usa aquí para dibujar una campana de forma similar a la t
# centrada en x_bar con la escala correcta dada por el error estándar estimado.
# Si quisiéramos dibujar la t exacta, sería centrada en 0 y la escala sería 1.
# La forma correcta de visualizar la distribución muestral de la media (que sigue una t
# cuando estandarizada) centrada en x_bar es usar una normal con la misma media y
# desviación estándar que la distribución muestral de la media.
# Vamos a dibujar una normal centrada en x_bar con el error estándar estimado para fines de visualización
y = norm.pdf(x, mu_distribucion, sigma_distribucion)


# --- Paso 3.2: Configurar la gráfica ---
plt.figure(figsize=(10, 6))
# Nota: Aunque usamos norm.pdf para dibujar la forma de la campana centrada en x_bar,
# la inferencia se basa en la distribución t debido a la muestra pequeña y sigma desconocida.
plt.plot(x, y, label='Distribución Muestral de la Media (Aproximada por Normal)', color='blue')
plt.title('Intervalo de Confianza del 95% para la Temperatura Promedio del Café (Distribución t)')
plt.xlabel('Temperatura (°C)')
plt.ylabel('Densidad de Probabilidad')
plt.grid(True)

# --- Paso 3.3: Rellenar el área del intervalo de confianza ---
x_fill = np.linspace(limite_inferior, limite_superior, 100)
y_fill = norm.pdf(x_fill, mu_distribucion, sigma_distribucion)
plt.fill_between(x_fill, y_fill, color='lightblue', alpha=0.6, label=f'Área del {nivel_confianza * 100}% de Confianza')

# --- Paso 3.4: Marcar la media muestral y los límites del intervalo ---
plt.axvline(x=x_bar, color='red', linestyle='--', label=f'Media Muestral ($\\bar{{x}}$) = {x_bar} °C')
plt.axvline(x=limite_inferior, color='green', linestyle=':', label=f'Límite Inferior = {limite_inferior:.2f} °C')
plt.axvline(x=limite_superior, color='green', linestyle=':', label=f'Límite Superior = {limite_superior:.2f} °C')

plt.legend()
plt.show()


# -----------------------------------------------------------------------------
# --- 4. Conclusión ---
# -----------------------------------------------------------------------------

print("\n--- Conclusión ---")
print(f"Con un nivel de confianza del 95%, se puede afirmar que la temperatura promedio")
print(f"de todas las tazas de café servidas por la cadena se encuentra en el rango")
print(f"de {limite_inferior:.2f} °C a {limite_superior:.2f} °C.")
print("Esto significa que si se repitiera el proceso de muestreo y cálculo de intervalos")
print("muchas veces, aproximadamente el 95% de esos intervalos contendrían la verdadera media poblacional.")
```

```python
# Importación de librerías necesarias
import numpy as np
from scipy.stats import t  # Importar la distribución t de Student
import matplotlib.pyplot as plt

# -----------------------------------------------------------------------------
# Problema 9: Intervalo de confianza para el tiempo promedio de entrega de pizza
# -----------------------------------------------------------------------------

# --- 1. Definición de los parámetros del problema ---
n = 20  # Tamaño de la muestra de entregas
x_bar = 30  # Tiempo promedio de entrega de la muestra (minutos)
s = 4  # Desviación estándar muestral (minutos)
nivel_confianza = 0.95  # Nivel de confianza deseado

# -----------------------------------------------------------------------------
# --- 2. Cálculo del Intervalo de Confianza ---
# -----------------------------------------------------------------------------

# La fórmula para el intervalo de confianza de la media poblacional (con sigma desconocida y n pequeño) es:
# IC = x_bar +/- t * (s / sqrt(n))

# --- Paso 2.1: Encontrar el valor crítico t ---
# El nivel de significancia (alfa) es 1 - nivel_confianza.
alfa = 1 - nivel_confianza
# Los grados de libertad para la distribución t son n - 1.
grados_libertad = n - 1
# Usamos scipy.stats.t.ppf() para encontrar el valor t correspondiente a la probabilidad (1 - alfa/2)
# con los grados de libertad adecuados.
t_critico = t.ppf(1 - alfa / 2, grados_libertad)

print(f"Nivel de confianza: {nivel_confianza * 100}%")
print(f"Nivel de significancia (alfa): {alfa}")
print(f"Grados de libertad: {grados_libertad}")
print(f"Valor crítico t para el {nivel_confianza * 100}% de confianza: {t_critico:.4f}")

# --- Paso 2.2: Calcular el error estándar estimado (s_x_bar) ---
# El error estándar estimado es la desviación estándar muestral dividida por la raíz cuadrada del tamaño de la muestra.
error_estandar_estimado = s / np.sqrt(n)

print(f"Error estándar estimado: {error_estandar_estimado:.4f}")

# --- Paso 2.3: Calcular el margen de error ---
# El margen de error es el valor crítico t multiplicado por el error estándar estimado.
margen_error = t_critico * error_estandar_estimado

print(f"Margen de error: {margen_error:.4f} minutos")

# --- Paso 2.4: Construir los límites del intervalo ---
# Límite inferior = media muestral - margen de error
limite_inferior = x_bar - margen_error
# Límite superior = media muestral + margen de error
limite_superior = x_bar + margen_error

print(f"\nIntervalo de confianza del {nivel_confianza * 100}%:")
print(f"[{limite_inferior:.2f} minutos, {limite_superior:.2f} minutos]")

# -----------------------------------------------------------------------------
# --- 3. Graficación de la Distribución ---
# -----------------------------------------------------------------------------

# Para la gráfica, usaremos la distribución t con grados de libertad n-1, centrada en x_bar
mu_distribucion = x_bar
sigma_distribucion = error_estandar_estimado

# Generar puntos en el eje x para la curva
x = np.linspace(mu_distribucion - 4 * sigma_distribucion, mu_distribucion + 4 * sigma_distribucion, 200)
# Calcular los valores de la función de densidad de probabilidad (PDF) usando la normal for visualization
y = norm.pdf(x, mu_distribucion, sigma_distribucion)


# --- Paso 3.2: Configurar la gráfica ---
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Distribución Muestral de la Media (Aproximada por Normal)', color='blue')
plt.title('Intervalo de Confianza del 95% para el Tiempo Promedio de Entrega de Pizza (Distribución t)')
plt.xlabel('Tiempo de Entrega (minutos)')
plt.ylabel('Densidad de Probabilidad')
plt.grid(True)

# --- Paso 3.3: Rellenar el área del intervalo de confianza ---
x_fill = np.linspace(limite_inferior, limite_superior, 100)
y_fill = norm.pdf(x_fill, mu_distribucion, sigma_distribucion)
plt.fill_between(x_fill, y_fill, color='lightblue', alpha=0.6, label=f'Área del {nivel_confianza * 100}% de Confianza')

# --- Paso 3.4: Marcar la media muestral y los límites del intervalo ---
plt.axvline(x=x_bar, color='red', linestyle='--', label=f'Media Muestral ($\\bar{{x}}$) = {x_bar} minutos')
plt.axvline(x=limite_inferior, color='green', linestyle=':', label=f'Límite Inferior = {limite_inferior:.2f} minutos')
plt.axvline(x=limite_superior, color='green', linestyle=':', label=f'Límite Superior = {limite_superior:.2f} minutos')

plt.legend()
plt.show()

# -----------------------------------------------------------------------------
# --- 4. Conclusión ---
# -----------------------------------------------------------------------------

print("\n--- Conclusión ---")
print(f"Con un nivel de confianza del {nivel_confianza * 100}%, podemos afirmar que el verdadero tiempo")
print(f"promedio de entrega de pizza se encuentra en el rango de {limite_inferior:.2f} minutos a {limite_superior:.2f} minutos.")
print("Esto significa que si se repitieran las muestras y se calcularan los intervalos de")
print("confianza de esta manera, el {nivel_confianza * 100}% de ellos contendrían la verdadera media poblacional.")
```

11. El departamento de salud planea una encuesta para estimar la proporción de residentes vacunados
contra la influenza. En un estudio piloto, el 72% de una muestra mostró estar vacunado. Determine
el tamaño de muestra necesario para estimar esta proporción con un margen de error del 3% y un
nivel de confianza del 95%. Si no se tuviera el estudio piloto, ¿cuál sería el tamaño requerido?

```python
import numpy as np
from scipy.stats import norm
import math

# -----------------------------------------------------------------------------
# Problema 11: Tamaño de muestra para estimar una proporción
# -----------------------------------------------------------------------------

# --- 1. Parámetros dados en el problema ---
margen_error_deseado = 0.03  # Margen de error deseado (3%)
nivel_confianza = 0.95  # Nivel de confianza deseado

# --- Escenario 1: Con estudio piloto ---
proporcion_estimada_piloto = 0.72  # Proporción estimada del estudio piloto (72%)

# --- Escenario 2: Sin estudio piloto ---
# En ausencia de una estimación previa, usamos p = 0.5 para maximizar el tamaño de muestra
proporcion_sin_piloto = 0.5

# -----------------------------------------------------------------------------
# --- 2. Cálculo del tamaño de muestra ---
# -----------------------------------------------------------------------------

# La fórmula para el tamaño de muestra (n) para estimar una proporción es:
# n = (Z^2 * p * (1-p)) / E^2
# donde Z es el valor crítico de la distribución normal estándar,
# p es la proporción estimada (o 0.5 si no hay estimación),
# y E es el margen de error deseado.

# --- Paso 2.1: Encontrar el valor crítico Z ---
# El nivel de significancia (alfa) es 1 - nivel_confianza.
alfa = 1 - nivel_confianza
# El valor Z_alfa/2 se obtiene de la distribución normal estándar.
z_critico = norm.ppf(1 - alfa / 2)

print(f"Nivel de confianza: {nivel_confianza * 100}%")
print(f"Margen de error deseado: {margen_error_deseado}")
print(f"Valor crítico Z: {z_critico:.4f}")

# --- Paso 2.2: Calcular el tamaño de muestra para el Escenario 1 (con estudio piloto) ---
p1 = proporcion_estimada_piloto
q1 = 1 - p1
n1 = (z_critico**2 * p1 * q1) / (margen_error_deseado**2)

# Redondear hacia arriba al entero más cercano, ya que el tamaño de muestra debe ser un número entero
n1_redondeado = math.ceil(n1)

print("\n--- Escenario 1: Con estudio piloto ---")
print(f"Proporción estimada (p): {p1}")
print(f"Tamaño de muestra calculado: {n1:.2f}")
print(f"Tamaño de muestra requerido (redondeado hacia arriba): {n1_redondeado}")

# --- Paso 2.3: Calcular el tamaño de muestra para el Escenario 2 (sin estudio piloto) ---
p2 = proporcion_sin_piloto
q2 = 1 - p2
n2 = (z_critico**2 * p2 * q2) / (margen_error_deseado**2)

# Redondear hacia arriba
n2_redondeado = math.ceil(n2)

print("\n--- Escenario 2: Sin estudio piloto ---")
print(f"Proporción supuesta (p): {p2} (para maximizar el tamaño)")
print(f"Tamaño de muestra calculado: {n2:.2f}")
print(f"Tamaño de muestra requerido (redondeado hacia arriba): {n2_redondeado}")

# -----------------------------------------------------------------------------
# --- 3. Conclusión ---
# -----------------------------------------------------------------------------

print("\n--- Conclusión ---")
print(f"Para estimar la proporción de residentes vacunados con un margen de error del")
print(f"{margen_error_deseado*100}% y un nivel de confianza del {nivel_confianza*100}%:")
print(f"- Si se utiliza la estimación del estudio piloto (p={proporcion_estimada_piloto}), se requiere un tamaño de muestra de al menos {n1_redondeado} residentes.")
print(f"- Si no se tiene una estimación previa y se utiliza p=0.5 para la máxima variabilidad, se requiere un tamaño de muestra de al menos {n2_redondeado} residentes.")
print("El tamaño de muestra es mayor cuando no se tiene una estimación previa,")
print("ya que p=0.5 representa la máxima incertidumbre en la proporción.")
```

```python
import numpy as np
from scipy.stats import t
import matplotlib.pyplot as plt

# -----------------------------------------------------------------------------
# Problema 26: Prueba de hipótesis para el tiempo de entrega (Prueba t)
# -----------------------------------------------------------------------------

# --- 1. Definir hipótesis y parámetros ---
# H0: mu = 25 minutos
# Ha: mu > 25 minutos (prueba de una cola a la derecha)
mu_0 = 25          # Media poblacional hipotética (minutos)
s = 4              # Desviación estándar muestral (minutos)
n = 15             # Tamaño de la muestra
x_bar = 28         # Media de la muestra (minutos)
alfa = 0.05        # Nivel de significancia

# -----------------------------------------------------------------------------
# --- 2. Cálculo del estadístico t ---
# -----------------------------------------------------------------------------
# Fórmula: t = (x_bar - mu_0) / (s / sqrt(n))
error_estandar = s / np.sqrt(n)
t_calculado = (x_bar - mu_0) / error_estandar

print(f"Estadístico de prueba t calculado: {t_calculado:.4f}")

# -----------------------------------------------------------------------------
# --- 3. Determinación de grados de libertad, valor crítico y región de rechazo ---
# -----------------------------------------------------------------------------
# Grados de libertad (df) = n - 1
df = n - 1

# El valor crítico para una prueba de cola a la derecha se encuentra en
# el percentil (1 - alfa) de la distribución t de Student.
t_critico = t.ppf(1 - alfa, df=df)

# El valor p se calcula como la probabilidad de obtener un valor t igual o más
# extremo que el calculado.
p_valor = 1 - t.cdf(t_calculado, df=df)

print(f"Grados de libertad (df): {df}")
print(f"Valor crítico t (para alfa = {alfa}): {t_critico:.4f}")
print(f"Valor p: {p_valor:.4f}")

# -----------------------------------------------------------------------------
# --- 4. Graficación de la distribución t de Student ---
# -----------------------------------------------------------------------------
x = np.linspace(-4, 4, 200)
y = t.pdf(x, df=df)

plt.figure(figsize=(10, 6))
plt.plot(x, y, color='blue', label=f'Distribución t de Student (df={df})')
plt.title('Región de Rechazo para una Prueba de una Cola a la Derecha')
plt.xlabel('Valor t')
plt.ylabel('Densidad de Probabilidad')
plt.grid(True)

# Graficar la región de rechazo (cola derecha)
x_rechazo = np.linspace(t_critico, 4, 100)
y_rechazo = t.pdf(x_rechazo, df=df)
plt.fill_between(x_rechazo, y_rechazo, color='red', alpha=0.6, label=f'Región de Rechazo (alfa = {alfa})')

# Marcar el valor crítico
plt.axvline(x=t_critico, color='red', linestyle='--', label=f'Valor Crítico t = {t_critico:.4f}')

# Marcar el estadístico de prueba calculado
plt.axvline(x=t_calculado, color='green', linestyle='-', linewidth=2, label=f'Estadístico t = {t_calculado:.4f}')

plt.legend()
plt.show()

# -----------------------------------------------------------------------------
# --- 5. Decisión y Conclusión ---
# -----------------------------------------------------------------------------
print("\n--- Decisión y Conclusión ---")
if t_calculado > t_critico:
    print(f"Decisión: El estadístico de prueba t ({t_calculado:.4f}) es mayor que el valor crítico ({t_critico:.4f}).")
    print("Por lo tanto, se RECHAZA la hipótesis nula (H0).")
    print(f"El valor p ({p_valor:.4f}) es menor que el nivel de significancia ({alfa}).")
    print("Se RECHAZA H0.")
    print("Conclusión práctica: Hay suficiente evidencia para afirmar que el tiempo promedio de entrega es mayor a 25 minutos.")
else:
    print(f"Decisión: El estadístico de prueba t ({t_calculado:.4f}) es menor que el valor crítico ({t_critico:.4f}).")
    print("Por lo tanto, NO se RECHAZA la hipótesis nula (H0).")
    print(f"El valor p ({p_valor:.4f}) es mayor que el nivel de significancia ({alfa}).")
    print("NO se RECHAZA H0.")
    print("Conclusión práctica: No hay suficiente evidencia para afirmar que el tiempo promedio de entrega es mayor a 25 minutos.")
```

```python
import numpy as np
from scipy.stats import chi2
import matplotlib.pyplot as plt

# -----------------------------------------------------------------------------
# Problema 31: Prueba de hipótesis para la varianza del tiempo de llamada (Prueba Chi-cuadrado)
# -----------------------------------------------------------------------------

# --- 1. Definir hipótesis y parámetros ---
# H0: sigma^2 = 2.5 minutos^2 (La varianza poblacional es 2.5 minutos^2)
# Ha: sigma^2 != 2.5 minutos^2 (La varianza poblacional es diferente de 2.5 minutos^2 - prueba de dos colas)
sigma2_0 = 2.5       # Varianza poblacional hipotética (minutos^2)
n = 16               # Tamaño de la muestra
s2 = 4               # Varianza muestral (minutos^2)
alfa = 0.05          # Nivel de significancia

print("--- Hipótesis ---")
print(f"Hipótesis Nula (H0): sigma^2 = {sigma2_0} minutos^2")
print(f"Hipótesis Alternativa (Ha): sigma^2 != {sigma2_0} minutos^2 (Prueba de dos colas)")

# -----------------------------------------------------------------------------
# --- 2. Cálculo del estadístico Chi-cuadrado ---
# -----------------------------------------------------------------------------
# Fórmula: Chi^2 = ((n - 1) * s^2) / sigma^2_0
chi2_calculado = ((n - 1) * s2) / sigma2_0

print(f"\nEstadístico de prueba Chi-cuadrado calculado: {chi2_calculado:.4f}")

# -----------------------------------------------------------------------------
# --- 3. Determinación de grados de libertad, valor crítico(s) y región(es) de rechazo ---
# -----------------------------------------------------------------------------
# Grados de libertad (df) = n - 1
df = n - 1

# Para una prueba de dos colas, el nivel de significancia se divide en dos.
# El valor crítico izquierdo se encuentra en el percentil alfa/2.
chi2_critico_izq = chi2.ppf(alfa / 2, df=df)
# El valor crítico derecho se encuentra en el percentil 1 - alfa/2.
chi2_critico_der = chi2.ppf(1 - alfa / 2, df=df)

# El valor p para una prueba de dos colas con un estadístico calculado es
# la suma de las probabilidades de las colas extremas.
# Si el estadístico está en la cola izquierda, el p-valor es 2 * P(Chi^2 <= chi2_calculado).
# Si el estadístico está en la cola derecha, el p-valor es 2 * P(Chi^2 >= chi2_calculado).
# En general, es 2 * min(P(Chi^2 <= chi2_calculado), P(Chi^2 >= chi2_calculado)).
p_valor = 2 * min(chi2.cdf(chi2_calculado, df=df), 1 - chi2.cdf(chi2_calculado, df=df))


print(f"Grados de libertad (df): {df}")
print(f"Valores críticos Chi-cuadrado (para alfa = {alfa}): Izquierdo = {chi2_critico_izq:.4f}, Derecho = {chi2_critico_der:.4f}")
print(f"Regiones de rechazo: Rechazar H0 si Chi^2_calculado < {chi2_critico_izq:.4f} o Chi^2_calculado > {chi2_critico_der:.4f}")
print(f"Valor p: {p_valor:.4f}")


# -----------------------------------------------------------------------------
# --- 4. Graficación de la distribución Chi-cuadrado ---
# -----------------------------------------------------------------------------
# Definir un rango de valores en el eje x para la gráfica
# La distribución Chi-cuadrado es asimétrica y no negativa
x = np.linspace(0, chi2.ppf(0.999, df=df), 200) # Ir hasta un percentil alto para capturar la cola
y = chi2.pdf(x, df=df)

plt.figure(figsize=(10, 6))
plt.plot(x, y, color='blue', label=f'Distribución Chi-cuadrado (df={df})')
plt.title('Regiones de Rechazo para una Prueba de Varianza de Dos Colas')
plt.xlabel('Valor Chi-cuadrado')
plt.ylabel('Densidad de Probabilidad')
plt.grid(True)

# Graficar las regiones de rechazo (ambas colas)
x_rechazo_izq = np.linspace(0, chi2_critico_izq, 100)
y_rechazo_izq = chi2.pdf(x_rechazo_izq, df=df)
plt.fill_between(x_rechazo_izq, y_rechazo_izq, color='red', alpha=0.6, label=f'Región de Rechazo (alfa/2 = {alfa/2})')

x_rechazo_der = np.linspace(chi2_critico_der, x[-1], 100) # Usar el último punto del rango x
y_rechazo_der = chi2.pdf(x_rechazo_der, df=df)
plt.fill_between(x_rechazo_der, y_rechazo_der, color='red', alpha=0.6)

# Marcar los valores críticos
plt.axvline(x=chi2_critico_izq, color='red', linestyle='--', label=f'Valores Críticos Chi^2 = {chi2_critico_izq:.4f}, {chi2_critico_der:.4f}')
plt.axvline(x=chi2_critico_der, color='red', linestyle='--')

# Marcar el estadístico de prueba calculado
plt.axvline(x=chi2_calculado, color='green', linestyle='-', linewidth=2, label=f'Estadístico Chi^2 = {chi2_calculado:.4f}')

plt.legend()
plt.xlim(0, x[-1]) # Asegurar que el eje x comience en 0
plt.show()


# -----------------------------------------------------------------------------
# --- 5. Decisión y Conclusión ---
# -----------------------------------------------------------------------------
print("\n--- Decisión y Conclusión ---")
if chi2_calculado < chi2_critico_izq or chi2_calculado > chi2_critico_der:
    print(f"Decisión: El estadístico de prueba Chi-cuadrado calculado ({chi2_calculado:.4f}) cae en la región de rechazo.")
    print("Por lo tanto, se RECHAZA la hipótesis nula (H0).")
    print(f"El valor p ({p_valor:.4f}) es menor que el nivel de significancia ({alfa}).")
    print("Se RECHAZA H0.")
    print("Conclusión práctica: Hay suficiente evidencia estadística, con un nivel de significancia de 0.05, para afirmar que la varianza poblacional del tiempo de duración de las llamadas difiere de 2.5 minutos^2.")
else:
    print(f"Decisión: El estadístico de prueba Chi-cuadrado calculado ({chi2_calculado:.4f}) NO cae en la región de rechazo.")
    print("Por lo tanto, NO se RECHAZA la hipótesis nula (H0).")
    print(f"El valor p ({p_valor:.4f}) es mayor que el nivel de significancia ({alfa}).")
    print("NO se RECHAZA H0.")
    print("Conclusión práctica: No hay suficiente evidencia estadística, con un nivel de significancia de 0.05, para afirmar que la varianza poblacional del tiempo de duración de las llamadas difiere de 2.5 minutos^2.")
```

32. Una máquina embotelladora debe mantener la varianza del llenado en 2 = 1,2 ml2. Se toma una
muestra de n = 20 botellas y se encuentra s2 = 2,0. ¿Existe evidencia al 1% de que la varianza es
mayor?
Resuelva siguiendo el siguiente protocolo:
Plantee las hipótesis nula y alternativa.
Calcule el estadístico 2.
Determine los grados de libertad y el valor crítico, además de la región crítica y la región de
rechazo.
Concluya si se rechaza H0.

```python
import numpy as np
from scipy.stats import chi2
import matplotlib.pyplot as plt

# -----------------------------------------------------------------------------
# Problema 32: Prueba de hipótesis para la varianza del llenado de botellas (Prueba Chi-cuadrado)
# -----------------------------------------------------------------------------

# --- 1. Definir hipótesis y parámetros ---
# H0: sigma^2 = 1.2 ml^2 (La varianza poblacional del llenado es 1.2 ml^2)
# Ha: sigma^2 > 1.2 ml^2 (La varianza poblacional del llenado es mayor a 1.2 ml^2 - prueba de una cola a la derecha)
sigma2_0 = 1.2       # Varianza poblacional hipotética (ml^2)
n = 20               # Tamaño de la muestra
s2 = 2.0             # Varianza muestral (ml^2)
alfa = 0.01          # Nivel de significancia

print("--- Hipótesis ---")
print(f"Hipótesis Nula (H0): sigma^2 = {sigma2_0} ml^2")
print(f"Hipótesis Alternativa (Ha): sigma^2 > {sigma2_0} ml^2 (Prueba de una cola a la derecha)")

# -----------------------------------------------------------------------------
# --- 2. Cálculo del estadístico Chi-cuadrado ---
# -----------------------------------------------------------------------------
# Fórmula: Chi^2 = ((n - 1) * s^2) / sigma^2_0
chi2_calculado = ((n - 1) * s2) / sigma2_0

print(f"\nEstadístico de prueba Chi-cuadrado calculado: {chi2_calculado:.4f}")

# -----------------------------------------------------------------------------
# --- 3. Determinación de grados de libertad, valor crítico y región de rechazo ---
# -----------------------------------------------------------------------------
# Grados de libertad (df) = n - 1
df = n - 1

# Para una prueba de una cola a la derecha, el valor crítico se encuentra en
# el percentil (1 - alfa) de la distribución Chi-cuadrado.
chi2_critico = chi2.ppf(1 - alfa, df=df)

# El valor p para una prueba de cola a la derecha es la probabilidad de obtener
# un valor Chi-cuadrado igual o más extremo (mayor) que el calculado.
p_valor = 1 - chi2.cdf(chi2_calculado, df=df)


print(f"Grados de libertad (df): {df}")
print(f"Valor crítico Chi-cuadrado (para alfa = {alfa}): {chi2_critico:.4f}")
print(f"Región de rechazo: Rechazar H0 si Chi^2_calculado > {chi2_critico:.4f}")
print(f"Valor p: {p_valor:.4f}")


# -----------------------------------------------------------------------------
# --- 4. Graficación de la distribución Chi-cuadrado ---
# -----------------------------------------------------------------------------
# Definir un rango de valores en el eje x para la gráfica
# La distribución Chi-cuadrado es asimétrica y no negativa
x = np.linspace(0, chi2.ppf(0.999, df=df), 200) # Ir hasta un percentil alto para capturar la cola
y = chi2.pdf(x, df=df)

plt.figure(figsize=(10, 6))
plt.plot(x, y, color='blue', label=f'Distribución Chi-cuadrado (df={df})')
plt.title('Región de Rechazo para una Prueba de Varianza de una Cola a la Derecha')
plt.xlabel('Valor Chi-cuadrado')
plt.ylabel('Densidad de Probabilidad')
plt.grid(True)

# Graficar la región de rechazo (cola derecha)
x_rechazo_der = np.linspace(chi2_critico, x[-1], 100) # Usar el último punto del rango x
y_rechazo_der = chi2.pdf(x_rechazo_der, df=df)
plt.fill_between(x_rechazo_der, y_rechazo_der, color='red', alpha=0.6, label=f'Región de Rechazo (alfa = {alfa})')

# Marcar el valor crítico
plt.axvline(x=chi2_critico, color='red', linestyle='--', label=f'Valor Crítico Chi^2 = {chi2_critico:.4f}')

# Marcar el estadístico de prueba calculado
plt.axvline(x=chi2_calculado, color='green', linestyle='-', linewidth=2, label=f'Estadístico Chi^2 = {chi2_calculado:.4f}')

plt.legend()
plt.xlim(0, x[-1]) # Asegurar que el eje x comience en 0
plt.show()


# -----------------------------------------------------------------------------
# --- 5. Decisión y Conclusión ---
# -----------------------------------------------------------------------------
print("\n--- Decisión y Conclusión ---")
if chi2_calculado > chi2_critico:
    print(f"Decisión: El estadístico de prueba Chi-cuadrado calculado ({chi2_calculado:.4f}) cae en la región de rechazo.")
    print("Por lo tanto, se RECHAZA la hipótesis nula (H0).")
    print(f"El valor p ({p_valor:.4f}) es menor que el nivel de significancia ({alfa}).")
    print("Se RECHAZA H0.")
    print("Conclusión práctica: Hay suficiente evidencia estadística, con un nivel de significancia de 0.01, para afirmar que la varianza poblacional del llenado de las botellas es mayor a 1.2 ml^2.")
else:
    print(f"Decisión: El estadístico de prueba Chi-cuadrado calculado ({chi2_calculado:.4f}) NO cae en la región de rechazo.")
    print("Por lo tanto, NO se RECHAZA la hipótesis nula (H0).")
    print(f"El valor p ({p_valor:.4f}) es mayor que el nivel de significancia ({alfa}).")
    print("NO se RECHAZA H0.")
    print("Conclusión práctica: No hay suficiente evidencia estadística, con un nivel de significancia de 0.01, para afirmar que la varianza poblacional del llenado de las botellas es mayor a 1.2 ml^2.")
```

Resuelva siguiendo el siguiente protocolo:
Plantee las hipótesis nula y alternativa.
Calcule el estadístico 2.
Determine los grados de libertad y el valor crítico, además de la región crítica y la región de
rechazo.

```python
import numpy as np
from scipy.stats import chi2
import matplotlib.pyplot as plt

# -----------------------------------------------------------------------------
# Problema 33: Prueba de hipótesis para la varianza del grosor de piezas metálicas (Prueba Chi-cuadrado)
# -----------------------------------------------------------------------------

# --- 1. Definir hipótesis y parámetros ---
# H0: sigma^2 >= 0.5 mm^2 (La varianza poblacional es al menos 0.5 mm^2)
# Ha: sigma^2 < 0.5 mm^2 (La varianza poblacional es menor a 0.5 mm^2 - prueba de una cola a la izquierda)
sigma2_0 = 0.5       # Varianza poblacional hipotética (mm^2)
n = 12               # Tamaño de la muestra
s2 = 0.3             # Varianza muestral (mm^2)
alfa = 0.05          # Nivel de significancia

print("--- Hipótesis ---")
print(f"Hipótesis Nula (H0): sigma^2 >= {sigma2_0} mm^2")
print(f"Hipótesis Alternativa (Ha): sigma^2 < {sigma2_0} mm^2 (Prueba de una cola a la izquierda)")

# -----------------------------------------------------------------------------
# --- 2. Cálculo del estadístico Chi-cuadrado ---
# -----------------------------------------------------------------------------
# Fórmula: Chi^2 = ((n - 1) * s^2) / sigma^2_0
chi2_calculado = ((n - 1) * s2) / sigma2_0

print(f"\nEstadístico de prueba Chi-cuadrado calculado: {chi2_calculado:.4f}")

# -----------------------------------------------------------------------------
# --- 3. Determinación de grados de libertad, valor crítico y región de rechazo ---
# -----------------------------------------------------------------------------
# Grados de libertad (df) = n - 1
df = n - 1

# Para una prueba de una cola a la izquierda, el valor crítico se encuentra en
# el percentil alfa de la distribución Chi-cuadrado.
chi2_critico = chi2.ppf(alfa, df=df)

# El valor p para una prueba de cola a la izquierda es la probabilidad de obtener
# un valor Chi-cuadrado igual o más extremo (menor) que el calculado.
p_valor = chi2.cdf(chi2_calculado, df=df)


print(f"Grados de libertad (df): {df}")
print(f"Valor crítico Chi-cuadrado (para alfa = {alfa}): {chi2_critico:.4f}")
print(f"Región de rechazo: Rechazar H0 si Chi^2_calculado < {chi2_critico:.4f}")
print(f"Valor p: {p_valor:.4f}")


# -----------------------------------------------------------------------------
# --- 4. Graficación de la distribución Chi-cuadrado ---
# -----------------------------------------------------------------------------
# Definir un rango de valores en el eje x para la gráfica
# La distribución Chi-cuadrado es asimétrica y no negativa.
# Empezamos en 0 y vamos hasta un valor que capture la mayor parte de la distribución.
x = np.linspace(0, chi2.ppf(0.99, df=df), 200)
y = chi2.pdf(x, df=df)

plt.figure(figsize=(10, 6))
plt.plot(x, y, color='blue', label=f'Distribución Chi-cuadrado (df={df})')
plt.title('Región de Rechazo para una Prueba de Varianza de una Cola a la Izquierda')
plt.xlabel('Valor Chi-cuadrado')
plt.ylabel('Densidad de Probabilidad')
plt.grid(True)

# Graficar la región de rechazo (cola izquierda)
x_rechazo_izq = np.linspace(0, chi2_critico, 100)
y_rechazo_izq = chi2.pdf(x_rechazo_izq, df=df)
plt.fill_between(x_rechazo_izq, y_rechazo_izq, color='red', alpha=0.6, label=f'Región de Rechazo (alfa = {alfa})')

# Marcar el valor crítico
plt.axvline(x=chi2_critico, color='red', linestyle='--', label=f'Valor Crítico Chi^2 = {chi2_critico:.4f}')

# Marcar el estadístico de prueba calculado
plt.axvline(x=chi2_calculado, color='green', linestyle='-', linewidth=2, label=f'Estadístico Chi^2 = {chi2_calculado:.4f}')

plt.legend()
plt.xlim(0, x[-1]) # Asegurar que el eje x comience en 0
plt.show()


# -----------------------------------------------------------------------------
# --- 5. Decisión y Conclusión ---
# -----------------------------------------------------------------------------
print("\n--- Decisión y Conclusión ---")
if chi2_calculado < chi2_critico:
    print(f"Decisión: El estadístico de prueba Chi-cuadrado calculado ({chi2_calculado:.4f}) cae en la región de rechazo.")
    print("Por lo tanto, se RECHAZA la hipótesis nula (H0).")
    print(f"El valor p ({p_valor:.4f}) es menor que el nivel de significancia ({alfa}).")
    print("Se RECHAZA H0.")
    print("Conclusión práctica: Hay suficiente evidencia estadística, con un nivel de significancia de 0.05, para afirmar que la varianza poblacional del grosor de las piezas metálicas es menor a 0.5 mm^2.")
else:
    print(f"Decisión: El estadístico de prueba Chi-cuadrado calculado ({chi2_calculado:.4f}) NO cae en la región de rechazo.")
    print("Por lo tanto, NO se RECHAZA la hipótesis nula (H0).")
    print(f"El valor p ({p_valor:.4f}) es mayor que el nivel de significancia ({alfa}).")
    print("NO se RECHAZA H0.")
    print("Conclusión práctica: No hay suficiente evidencia estadística, con un nivel de significancia de 0.05, para afirmar que la varianza poblacional del grosor de las piezas metálicas es menor a 0.5 mm^2.")
```
